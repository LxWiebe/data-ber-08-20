```python
import pandas as pd
import numpy as np
import re
import datetime
```

```python
data = pd.read_csv('GSAF5.csv', engine = 'python')
data.head()
```

# Handling missing values


```python
missing_values = data.isna().sum()
round(missing_values / len(data), 2)
```

#### Deleting columns: Time, Unnamed:22, Unnamed: 23 --> missing more than 50% of values

```python
data = data.drop(['Time', 'Unnamed: 22', 'Unnamed: 23'], axis = 1)
data.head()
```

# Column Country

```python
data.Country.unique()
```

```python
# not all entries are case sensitive
```

```python
data['Country'] = data['Country'].str.upper() 
```

# Column name

```python
data.Name.unique()
```

```python
data.Name.unique()
```

```python
# Column contains gender information --> remove values from column
```

```python
data['Name'] = data['Name'].replace(['male'], '').replace(['female'], '')
```

# Column Sex

```python
# Column name contains white Space -> Rename column name to remove white space
```

```python
data = data.rename(columns={'Sex ': 'Sex'})
```

```python
data.Sex.unique()
# Data not consistent --> remove values which are not F or M and remove whitespace from 'M '
```

```python
data['Sex'] = data['Sex'].replace(['lli'], None).replace(['N'], None).replace(['M '], 'M').replace(['M '], 'M').replace(['.'], None)
```

```python
data.Sex.unique()
```

# Column Age

```python
# Age contains not only numbers --> transform type of column into int and drop everything which doens´t contain numbers
```

```python
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
```

```python
data.dtypes
```

# Dropping Columns

```python
# Column Case Number is 3 x in data set
# Column Equal from column comparison is not need anymore
# --> drop these columns
```

```python
data.drop(['Case Number.1', 'Case Number.2', 'Equal'], axis=1)
```
