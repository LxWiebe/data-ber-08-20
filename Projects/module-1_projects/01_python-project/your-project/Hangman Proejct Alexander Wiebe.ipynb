```python
# Import random module
import random
```

```python
# Define word list, which the player has to guess
word_list =[
    'bonsai',
    'iron',
    'hack',
    'festival',
    'extend',
    'brown',
    'amateur',
    'hyped',
    'expert'
    'coding',
    'fridge',
    'horse',
    'reduce',
    'yellow',
    'finance',
    'energy',
    'humanity']
```

```python
# Call a word from word list and convert it to upper cases to make the comparison easier
def get_word():
    word = random.choice(word_list)
    return word.upper()
```

```python
# Create variables which are updated during the game

def game(word):
    word_completion = "_" * len(word) #The word initially contains only underscores
    guessed = False
    guessed_letters = [] # Holds the letters the user guessed
    guessed_words = [] # Holds the words the user guessed
    tries = 6 
    # Number of parts of the hangman (head, body, hands and feets) which corresponds to the tries the player
    # has before she loses
    print("Let's play Hangman!")
    print(display_hangman(tries))
    print(word_completion)
    print("\n")

# Define play loop
    while not guessed and tries > 0:
        guess = input("Please guess the word ar a letter: ").upper()

# guess.isalpha makes sure that the letter is from the alphabet

        if len(guess) == 1 and guess.isalpha():

# makes sure that play takes a new letter:
            if guess in guessed_letters:
                print("You already guessed the letter", guess)
            
# if player chooses wrong word, print a message and store the guessed letter into a list
            elif guess not in word:
                print(guess, "is not in the word.")
                tries -= 1
                guessed_letters.append(guess)

# if player guesses right
            else:
                print("Bingo!", guess, "is in the word!")
                guessed_letters.append(guess)

# indexing the searched word for replacement with the (correctly guessed letter)
                word_as_list = list(word_completion)
                indices = [i for i, letter in enumerate(word) if letter == guess]

# replaces underscore with the right guessed word
                for index in indices:
                    word_as_list[index] = guess
                word_completion = "".join(word_as_list)
            
# Checking wheather word is already completed
                if "_" not in word_completion:
                    guessed = True
        elif len(guess) == len(word) and guess.isalpha():
            if guess in guessed_words:
                print("You already guessed the word", guess)

# Flow for wrong answer
            elif guess != word:
                print(guess, "is not the word.")
                tries -= 1
                guessed_words.append(guess)
# Flo for right answer
            else:
                guessed = True
                word_completion = word
        else:
            print("Not a valid guess.")
# Visualize hangman
        print(display_hangman(tries))
        print(word_completion)
        print("\n")

# Final answers:
    if guessed:
        print("Yeah! You won")
    else:
        print("You lost. The man is hanged. The word was " + word)
```

```python
# Function that shows the hangman in each six stages

def display_hangman(tries):
    stages = [  # final state: head, torso, both arms, and both legs
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / \
                   -
                """,
                # head, torso, both arms, and one leg
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / 
                   -
                """,
                # head, torso, and both arms
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |      
                   -
                """,
                # head, torso, and one arm
                """
                   --------
                   |      |
                   |      O
                   |     \|
                   |      |
                   |     
                   -
                """,
                # head and torso
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                # head
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                # initial empty state
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return stages[tries]
```

```python
# definein main function which puts all functions into one

def main():
    word = get_word()
    game(word)
    while input("Play Again? (Y/N) ").upper() == "Y":
        word = get_word()
        game(word)
main()
```
